```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None
```

```python
df = pd.read_csv(r'C:\Users\sumsh\OneDrive\Desktop\mini projects\python correlation\movies.csv')
```

```python
df.head()
```

```python
#find missing data
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
#drop missing data
df = df.dropna()
```

```python
#data types of the columns 
df.dtypes
```

```python
#change data type 
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
df['votes'] = df['votes'].astype('int64')
df
```

```python
#create new year column based on the release date
df['yearcorrect'] = df['released'].astype(str).str[:4]
df
```

```python
df.sort_values(by = ['gross'], inplace = False, ascending = False)
```

```python
#if you want to see all the data
pd.set_option('display.max_row', None)
df.head()
```

```python
#find and drop duplicate
duplicate = df[df.duplicated()] 
duplicate
```

```python
# find gross revenue correlation 
#budget vs gross, company vs gross

plt.scatter(x = df['budget'], y = df['gross'])
plt.title('Budget vs Gross Earnings')
plt.ylabel('Budget')
plt.xlabel('gross Earnings')
plt.show
```

```python
df.head()
```

```python
#Plot budget vs gross using seaborn
sns.regplot( x = 'budget', y = 'gross', data = df, scatter_kws={"color" : "red"}, line_kws={'color': 'blue'})
```

```python
# df.corr only works with numerical values
pe = df.corr(method = 'pearson') #pearson, kendall, spearman
ken = df.corr(method = 'kendall')
pe
```

```python
ken
```

```python
# hugh correlation b/w budget and gross

corr_matrix = df.corr(method = 'pearson')
sns.heatmap(corr_matrix, annot = True)
plt.title('Correlation Matrix for Numeric Featurs')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
# look at company
#company is not numeric, creating numeric represeentations for the comapany

df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized.head()
```

```python
corr_matrix = df_numerized.corr(method = 'pearson')
sns.heatmap(corr_matrix, annot = True)
plt.title('Correlation Matrix for Numeric Featurs')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```
